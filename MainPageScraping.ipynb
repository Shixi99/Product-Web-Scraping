```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bfsp
```

```python
response = requests.get("https://tap.az/")
html_content = bfsp(response.content, "html.parser")
html_content
```

```python
names, prices, dates, links = [],[],[],[]

for product in html_content.findAll('a',{'class':'products-link'}):
    link = 'https://tap.az' + product.get('href')
    price = product.find('div', {'class': 'products-price'}).text
    name = product.find('div', {'class': 'products-name'}).text
    date = product.find('div', {'class': 'products-created'}).text

    names.append(name)
    prices.append(price)
    dates.append(date)
    links.append(link)
```

```python
data = list(zip(names, prices, dates, links))
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
col_names = ['Name','Price','Date','Link']
df = pd.DataFrame(data, columns=col_names)
df
```

```python
test_data = df
```

```python
for i in range(len(test_data["Date"])):
    test_data["Date"][i]=test_data["Date"][i].replace("dünən,","dünən")
```

```python
test_data
```

```python
test_data[["City","Date-Time"]] = test_data["Date"].str.split(",", expand=True)
```

```python
test_data.drop("Date", axis=1, inplace=True)
```

```python
test_data
```

```python

currency = ["AZN","USD","EUR"]

for i in range(len(test_data["Price"])):
        if any(substring in test_data["Price"][i] for substring in currency) == True:
            test_data["Price"][i]=test_data["Price"][i].replace(test_data["Price"][i][-3:],"-"+test_data["Price"][i][-3:])
```
