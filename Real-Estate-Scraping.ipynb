```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://bina.az/")
html_content = BeautifulSoup(response.content, "html.parser")
html_content
```

```python
names, prices, dates, links, locations, currencies = [],[],[],[],[],[]

for add in html_content.findAll('div',{'class':'items-i vipped'}):
    link = 'https://bina.az/' + add.find('a').get('href')
    price = add.find('span', {'class': 'price-val'}).get_text()
    currency = add.find('span',{'class':'price-cur'}).get_text()
    name = add.find('ul', {'class': 'name'}).get_text()
    location = add.find('div', {'class':'location'}).get_text()
    city_date_time = add.find('div',{'class':'city_when'}).get_text()

    names.append(name)
    prices.append(price)
    dates.append(city_date_time)
    links.append(link)
    locations.append(location)
    currencies.append(currency)
```

```python
data = list(zip(names, prices, currencies, dates, locations,links ))
data
```

```python
col_names = ['room_size', 'prince','currency','city-date-time','location','link']
df = pd.DataFrame(data, columns=col_names)
```

```python
df
```

```python
df[["City","Date-Time"]] = df["city-date-time"].str.split(",", expand=True)
```

```python
df.drop('city-date-time',axis=1, inplace=True)
```

```python
df[["num_of_rooms_size_of_apartment(m²)","floor"]] = df["room_size"].str.split('m²', expand=True)
```

```python
otaq = ['otaqlı']
for i in range(len(df["num_of_rooms_size_of_apartment(m²)"])):
        if any(substring in df["num_of_rooms_size_of_apartment(m²)"][i] for substring in otaq) == True:
            df["num_of_rooms_size_of_apartment(m²)"][i]=df["num_of_rooms_size_of_apartment(m²)"][i].replace(df["num_of_rooms_size_of_apartment(m²)"][i][:8], df["num_of_rooms_size_of_apartment(m²)"][i][:8]+'-')
```
